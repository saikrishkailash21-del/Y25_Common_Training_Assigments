## 4. Image Transformations
**Task:**
1. Create a function `translate(img, x, y)` to shift an image.
2. Create a function `rotate(img, angle, rotPoint=None)` to rotate an image.
3. Apply a flip transformation to `park.jpg`.

```python
# Write your code here
import cv2
import numpy as np
from matplotlib import pyplot as plt

def translate(img,x,y):
    M = np.float32([[1,0,x],[0,1,y]])
    (height,width) = img.shape[:2]
    return cv2.warpAffine(img,M,(width,height))

def rotate(img,angle,rotPoint = None):
    (height,width) = img.shape[:2]
    if rotPoint is None:
        rotPoint = (width//2,height//2)
    R = cv2.getRotationMatrix2D(rotPoint,angle,1)
    return cv2.warpAffine(img,R,(width,height))

img = cv2.imread('./opencv_files/park.jpg')
img1 = cv2.flip(img,-1)
plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB))
plt.plot()
```

## 5. Contour Detection

3. Find the edges using Canny.
4. Use `cv.findContours` to find all contours.
5. Print the count and draw them on a blank image.

```python
# Write your code here
import cv2
import matplotlib.pyplot as plt 
img = cv2.imread('./opencv_files/cat.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
gaussian = cv2.GaussianBlur(gray,(7,7),0)
edge = cv2.Canny(gaussian,100,200)
contours,heirachy = cv2.findContours(edge,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
image  = np.zeros((500,500,3),dtype = 'uint8')
cv2.drawContours(image,contours,-1,(0,255,0),2)

plt.imshow(image,cmap='gray')
plt.show()
```

## 7. Smoothing and Blurring
**Task:**
1. Load `cats.jpg`.
2. Apply Averaging, Gaussian and Median Blurs

```python
# Write your code here
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('./opencv_files/cat.jpg')
kernel = np.ones((3,3),np.float32)/9
avg = cv2.filter2D(img,-1,kernel)
gaussian = cv2.GaussianBlur(img,(5,5),0)
median =  cv2.medianBlur(img,5)

plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.imshow(img)
plt.title("Original")

plt.subplot(2,2,2)
plt.imshow(avg)
plt.title("Average Blur")

plt.subplot(2,2,3)
plt.imshow(gaussian)
plt.title("Gaussian Blur")

plt.subplot(2,2,4)
plt.imshow(median)
plt.title("Median Blur")

plt.show()
```

## 8. Bitwise Operations
**Task:**
1. Create a $400 \times 400$ blank image.
2. Create a rectangle and a circle on separate blank images.
3. Perform Bitwise AND, OR, XOR, and NOT.

```python
# Write your code here
import cv2
import numpy as np
img1 = np.zeros((400,400,3),dtype = 'uint8')
img2 = np.zeros((400,400,3),dtype = 'uint8')
img1[100:300,100:300] = (0,0,255)
cv2.circle(img2,(200,200),100,(0,0,255),-1)
bitwise_and = cv2.bitwise_and(img1,img2)
bitwise_or = cv2.bitwise_or(img1,img2)
bitwise_xor = cv2.bitwise_xor(img1,img2)
bitwise_not = cv2.bitwise_not(img1)
import matplotlib.pyplot as plt
plt.figure(figsize = (12,4))
plt.subplot(1,5,1)
plt.imshow(img1)
plt.title("rect")
plt.subplot(1,5,2)
plt.imshow(img2)
plt.title("circle")
plt.subplot(1,5,3)
plt.imshow(bitwise_and)
plt.title("bitwise and")
plt.subplot(1,5,4)
plt.imshow(bitwise_or)
plt.title("bitwise or")
plt.subplot(1,5,5)
plt.imshow(bitwise_xor)
plt.title("bitwise_xor")
plt.show()
```

## 9. Thresholding
**Task:**
1. Load `cats.jpg` and convert to grayscale.
2. Apply Simple Thresholding (Binary and Inverse).

```python
# Write your code here
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('./opencv_files/cat.jpg')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,binthresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
ret,invthresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY_INV)
plt.figure(figsize = (12,4))

plt.subplot(1,3,1)
plt.imshow(img)
plt.title("orginal")

plt.subplot(1,3,2)
plt.imshow(binthresh,cmap = 'gray')
plt.title("Binary")

plt.subplot(1,3,3)
plt.imshow(invthresh,cmap = 'gray')
plt.title("Inverse")

plt.show()
```
